```python
'''
corr2: 
parameters:
    data: np.array(4800*4000)
return:
    corr_df:逐行反复计算0时刻起至今的序列相关性
'''
```

```python
'''Python代码'''
def corr_py(df, n=30):
    l = []
    for i in range(0, n):
        px = df.iloc[0:i+1, :]
        corr = px.corrwith(px.mean(axis=1))
        l.append(corr)
    dff = pd.concat(l, axis=1).T
    return dff
```

```python
result = corr_py(data.iloc[0:100, :])
```
